# <b>Fourth Stage: Data Configuration</b>

<div style='text-align: justify'>The final dataset generated in the previous step stores flux combinations in list format. Before proceeding to model analysis, the data must be restructured so that each enzyme has its own dedicated column.</div>

```python
# Just pandas will be used for data manipulation
import pandas as pd
```

```python
# Load the final dataset
df_naive = pd.read_csv('../Datasets/final_reunite.csv')
display(df_naive)
```

```python
# List of reactions 
R_imp_final = ['PDH_fw','PPC_fw','CS_fw','ACONTa_fw','ACONTa_bw',
               'ACONTb_fw','ACONTb_bw','ICDHyr_fw','ICDHyr_bw',
               'AKGDH_fw','SUCOAS_fw','SUCOAS_bw','FUM_fw','FUM_bw',
               'FRD2_fw','ICL_fw','ICL_bw','MALS_fw']
```

```python
# list_aux will store the combinations in float format, as they are currently in string format

list_aux = []
for i in df_naive['Combination']:
    replacements = str.maketrans({"[": "", "]": ""})
    res = i.translate(replacements)
    list_aux.append(res.split(','))

for i in range(len(list_aux)):
    for j in range(len(R_imp_final)):
        list_aux[i][j] = float(list_aux[i][j])

list_aux
```

```python
"""Now, with the flux values converted to float format, a new data frame will be created containing only these values, excluding 
the biomass and NADH production columns.""" 

# The dictionary keys will be defined as numerical values, as this is more convenient for manipulation than using enzyme names.
dict_df = {k:[] for k in range(len(R_imp_final))}
for i in range(len(list_aux)):
    for j in range(len(list_aux[0])):
        dict_df[j].append(list_aux[i][j])

# The dictionary keys are reassigned to the corresponding enzyme names.
final_dict = dict(zip(R_imp_final, list(dict_df.values())))

# The data frame is created based on the previously constructed dictionary.
df_fluxes = pd.DataFrame(final_dict)
display(df_fluxes)
```

```python
# The flux values must now be merged with the biomass and NADH columns.

# It is also necessary to convert the NADH and biomass values to float format, as they were stored as strings.
df_fluxes['NADH Production'] = df_naive['NADH Production'].apply(lambda x: float(x))
df_fluxes['Biomass'] = df_naive['Biomass'].apply(lambda x: float(x))

display(df_fluxes)
```
