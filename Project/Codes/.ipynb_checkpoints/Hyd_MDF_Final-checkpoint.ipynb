```python
file_name = 'Hyd_Final_MDF (Cópia).tsv'
sbtab_file = open(file_name, 'r')
file_content = sbtab_file.read()
sbtab_file.close()


Sd = SBtab.SBtabDocument()
Sd.set_filename(file_name)
Sd.set_name('My_SBtab_Document')
Sd.add_sbtab_string(file_content, file_name)

for t in Sd.sbtabs:
    print(t.table_name)
    display(t.to_data_frame())
```

```python
pp_pfor = ThermodynamicModel.from_sbtab('Hyd_Final_MDF (Cópia).tsv')
pp_pfor.update_standard_dgs()

display (pp_pfor.net_reaction_formula)
```

```python
#pp_pfor.set_bounds('H2', Q_(4.28, 'mM'), Q_(4.28, 'mM'))
#pp_pfor.set_bounds('Nicotinamide_adenine_dinucleotide_reduced', Q_(150, 'mM'), Q_(150, 'mM'))
mdf_result_pfor = pp_pfor.mdf_analysis()
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
mdf_result_pfor.plot_driving_forces(ax=ax)

plt.savefig('MDF_Hyd.png',dpi=1200)
```

```python
def max_concentration(pp):
    suc_concentration = 0
    pp.set_bounds('H2', Q_(suc_concentration, 'mM'), Q_(suc_concentration, 'mM'))
    result = pp.mdf_analysis().score
    
    while True:
        suc_concentration += 0.1
        pp.set_bounds('H2', Q_(suc_concentration, 'mM'), Q_(suc_concentration, 'mM'))
        result = pp.mdf_analysis().score
        if result > 0 and result < 0.1:
            break
    return (suc_concentration,result)
```

```python
print(f"Maximum concentration of H2 produced: {max_concentration(pp_pfor)[0]} mM")
print(f"MDF value: {max_concentration(pp_pfor)[1]} kj/mol")
```
